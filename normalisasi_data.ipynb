```python
#Normalisasi Data (Standar Scaler)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
filelocation = 'C:/Users/User/Documents/Berkas Fikar/Skripsi Fikar/Database/data_gabungan_sekolah_sulsel.csv'
data = pd.read_csv(filelocation)
```

```python
data.head()
```

```python
x = data.loc[:, ['siswa_total', 'jumlah_guru', 'staf_total', 'jumlah_ruangan', 'daya_listrik', 'wilayah_total', 'skor_akreditasi', 'Lintang', 'Bujur']]
```

```python
scaler = StandardScaler()
scaler.fit (x)
data_scaled = scaler.transform(x)
```

```python
data_scaled
```

```python
data_scaled = pd.DataFrame(data_scaled, columns = ['siswa_total', 'jumlah_guru', 'staf_total', 'jumlah_ruangan', 'daya_listrik', 'wilayah_total', 'skor_akreditasi', 'Lintang', 'Bujur'])
```

```python
data_scaled
```

```python
data_scaled.to_csv('data_normalisasi_sekolah_sulsel.csv', index = False)
```
