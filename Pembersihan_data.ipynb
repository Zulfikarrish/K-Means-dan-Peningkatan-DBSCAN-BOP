```python
#Pembersihan Data

import pandas as pd
```

```python
filelocation = 'C:/Users/User/Documents/Berkas Fikar/Skripsi Fikar/Data/data_gabungan_sekolah_sulsel.csv'
df = pd.read_csv(filelocation)
```

```python
df.head()
```

```python
columns_to_clean = ['jumlah_ruangan', 'daya_listrik', 'wilayah_total', 'Lintang', 'Bujur']

def clean_data(value):
    if pd.isnull(value):  
        return value
    value = str(value).lower()
    value = value.replace('watt', '').replace('m2', '').replace('-', '').replace('.', '')
    try:
        return pd.to_numeric(value.strip())
    except ValueError:
        return pd.NA


df[columns_to_clean] = df[columns_to_clean].applymap(clean_data)
```
