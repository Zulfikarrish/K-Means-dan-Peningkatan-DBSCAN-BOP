```python
#Penerapan Teknik PCA

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
filelocation = 'C:/Users/User/Documents/Berkas Fikar/Skripsi Fikar/Database/data_normalisasi_sekolah_sulsel.csv'
data = pd.read_csv(filelocation)
```

```python
data.head()
```

```python
features = ['siswa_total', 'jumlah_guru', 'staf_total', 'jumlah_ruangan', 'daya_listrik', 'wilayah_total', 'skor_akreditasi', 'Lintang', 'Bujur']
x = data.loc[:, features].values
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
data_pca = pd.DataFrame(data = principalComponents, columns = ['pca1', 'pca2'])
```

```python
data_pca
```

```python
data_pca.to_csv('data_pca_sekolah_sulsel.csv', index = False)
```
